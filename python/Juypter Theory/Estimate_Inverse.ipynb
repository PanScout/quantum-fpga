```python
import numpy as np
import matplotlib.pyplot as plt

def newton_inverse(A, num_iterations=10, graph_error=False):
    """
    Compute the inverse of matrix A using Newton's method (Newton-Schulz iteration)
    and optionally graph the error at each iteration, using the highest precision
    (np.longdouble) available.

    Parameters:
        A (numpy.ndarray): A square invertible matrix.
        num_iterations (int): Number of iterations to perform.
        graph_error (bool): If True, plot the error vs. iteration.

    Returns:
        numpy.ndarray: An approximation to the inverse of A.
    """
    # Convert A to an array with highest precision (np.longdouble)
    A = np.array(A, dtype=np.longdouble)

    # Check if A is square
    n, m = A.shape
    if n != m:
        raise ValueError("Input matrix A must be square.")
    
    # Create the identity matrix with highest precision
    I = np.eye(n, dtype=np.longdouble)
    
    # Initial guess for the inverse: X0 = A^T / (||A||_1 * ||A||_inf)
    normA1 = np.linalg.norm(A, 1)
    normAinf = np.linalg.norm(A, np.inf)
    X = A.T / (normA1 * normAinf)
    
    # List to store error at each iteration
    errors = []
    
    # Compute and store the initial error norm
    initial_error = np.linalg.norm(I - A @ X)
    errors.append(initial_error)
    
    print("Initial guess X0:")
    print(X)
    print("Initial error norm ||I - A*X||:", initial_error)
    print("-" * 50)
    
    # Newton-Schulz iteration: X_{k+1} = X_k (2I - A X_k)
    for k in range(num_iterations):
        # Compute A * X_k
        AX = A @ X
        
        # Compute the correction term: 2I - A * X_k
        correction = 2 * I - AX
        
        print(f"Iteration {k+1}:")
        print("A * X:")
        print(AX)
        print("Correction (2I - A * X):")
        print(correction)
        
        # Update the approximation
        X = X @ correction
        
        # Compute the current error norm ||I - A*X||
        error = np.linalg.norm(I - A @ X)
        errors.append(error)
        
        print("Updated X:")
        print(X)
        print(f"Error norm ||I - A*X||: {error}")
        print("-" * 50)
    
    # Graph the error if requested
    if graph_error:
        plt.figure(figsize=(8, 5))
        plt.plot(range(num_iterations + 1), errors, marker='o', linestyle='-', color='b', label='Error Norm')
        plt.axhline(0, color='r', linestyle='--', label='Zero Error')
        plt.xlabel("Iteration")
        plt.ylabel("Error norm ||I - A*X||")
        plt.title("Error Norm vs. Iteration")
        plt.grid(True)
        # Uncomment the following line if you prefer a logarithmic scale for the error:
        plt.yscale("log")
        plt.legend()
        plt.show()
    
    return X

# Example usage:
if __name__ == "__main__":
    A = np.array(result.tolist(),dtype=float)
    # Compute its inverse using 50 iterations of Newton's method and graph the error
    invA = newton_inverse(A, num_iterations=50, graph_error=True)
    
    print("Approximate inverse of A:")
    print(invA)
    print("Check A * invA (should be close to the identity matrix):")
    print(A @ invA)
```

```python
import torch
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
import matplotlib.pyplot as plt

# Set device: use MPS if available on your M1 Mac, else fallback to CPU.
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print("Using device:", device)

def newton_inverse_torch(A, num_iterations=10, graph_error=False):
    """
    Compute the inverse of matrix A using the Newton-Schulz iteration on the GPU
    with PyTorch. The error at each iteration is computed and (optionally) plotted.
    
    Parameters:
        A (torch.Tensor): A square invertible matrix (assumed to be float32).
        num_iterations (int): Number of iterations to perform.
        graph_error (bool): If True, plot the error vs. iteration.
        
    Returns:
        X (torch.Tensor): Approximation to the inverse of A (on the same device as A).
        errors (list): List of error norms at each iteration (converted to Python floats).
    """
    n, m = A.shape
    if n != m:
        raise ValueError("Input matrix A must be square.")
    
    # Create identity matrix on the same device.
    I = torch.eye(n, device=device, dtype=A.dtype)
    
    # Compute norms for the initial guess.
    normA1 = torch.linalg.norm(A, 1)
    normAinf = torch.linalg.norm(A, float('inf'))
    
    # Initial guess: X0 = A^T / (||A||_1 * ||A||_inf)
    X = A.t() / (normA1 * normAinf)
    
    errors = []
    err = torch.linalg.norm(I - A @ X)
    errors.append(err.item())
    
    for k in range(num_iterations):
        AX = A @ X
        correction = 2 * I - AX
        X = X @ correction
        err = torch.linalg.norm(I - A @ X)
        errors.append(err.item())
    
    if graph_error:
        plt.figure(figsize=(8, 5))
        plt.plot(range(num_iterations + 1), errors, marker='o', linestyle='-', color='b')
        plt.axhline(0, color='r', linestyle='--')
        plt.xlabel("Iteration")
        plt.ylabel("Error norm ||I - A*X||")
        plt.title("Error Norm vs. Iteration (GPU, PyTorch)")
        plt.grid(True)
        plt.yscale("log")
        plt.show()
    
    return X, errors

def compute_errors_for_size(size, num_iterations, num_runs):
    """
    Compute the average error curve (over multiple runs) for a given matrix size,
    performing computations on the GPU.
    """
    errors_runs = []
    for _ in range(num_runs):
        # Create a random matrix A on the GPU.
        # Use torch.ceil and a scaling factor of 1E9 as in your original code.
        A = torch.ceil(1E9 * torch.rand(size, size, device=device, dtype=torch.float32))
        # Run the Newton-Schulz iteration (without graphing)
        _, errors = newton_inverse_torch(A, num_iterations=num_iterations, graph_error=False)
        errors_runs.append(errors)
    # Compute the elementwise mean over runs.
    return np.mean(errors_runs, axis=0)

# -------------------------------------------------------------------
# Parameters and Data Generation
# -------------------------------------------------------------------
# List of matrix sizes (you can adjust as needed)
#n_list = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 1536, 2048, 3072, 4096, 6_144, 8192]
n_list = [4096]
max_iterations = 50
num_runs = 8  # Number of random matrices per size

# Option: Use joblib to parallelize across different matrix sizes.
# If you run into issues with parallel GPU processing on M1, try setting n_jobs=1.
results = Parallel(n_jobs=-1)(
    delayed(compute_errors_for_size)(size, max_iterations, num_runs) for size in n_list
)

# Build a list of dictionaries (one per iteration per matrix size)
data = [
    {'iteration': i, 'size': size, 'error': errors_mean[i]}
    for size, errors_mean in zip(n_list, results)
    for i in range(len(errors_mean))
]

df = pd.DataFrame(data)
print(df.head())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# -----------------------------------------------------------------------------
# 1) Compute the iteration AFTER the maximum absolute difference between consecutive error values.
# -----------------------------------------------------------------------------
max_diff_data = []

for size, group in df.groupby('size'):
    group = group.sort_values('iteration')
    errors = group['error'].values
    iterations = group['iteration'].values
    # Compute the absolute differences between successive error values.
    diffs = np.abs(np.diff(errors))
    # The difference at index k corresponds to the change from iteration k to k+1.
    max_idx = np.argmax(diffs)
    # Record the iteration AFTER the maximum difference.
    # If max difference occurs between iterations k and k+1,
    # then we record iteration (k+1)+1 = k+2.
    max_diff_iter = iterations[max_idx + 1] + 1  
    max_diff_data.append({'size': size, 'max_diff_iter': max_diff_iter})

max_diff_df = pd.DataFrame(max_diff_data)
print("Iteration AFTER maximum error difference for each matrix size:")
print(max_diff_df)

# -----------------------------------------------------------------------------
# 2) Prepare Data for Modeling
# -----------------------------------------------------------------------------
X = max_diff_df[['size']].values   # Independent variable: matrix size
y = max_diff_df['max_diff_iter'].values   # Dependent variable: iteration after max diff

# -----------------------------------------------------------------------------
# 3) Define Fractional Polynomial Transformation
# -----------------------------------------------------------------------------
# Here we choose three fractional exponents for a degree-3 model.
# You can experiment with other fractional powers as needed.
fractional_powers = [1/2, 1]

def fractional_poly_transform(X, powers):
    """
    Transforms the input data X by raising it to each power in the list 'powers'.
    
    Parameters:
        X (array-like): Input data (n_samples x 1).
        powers (list): List of exponents to apply.
        
    Returns:
        A numpy array with columns corresponding to X raised to each power.
    """
    X = np.array(X).reshape(-1, 1)
    # Compute each fractional power and horizontally stack the results.
    transformed_features = np.hstack([np.power(X, p) for p in powers])
    return transformed_features

# Transform the independent variable using fractional powers.
X_frac = fractional_poly_transform(X, fractional_powers)

# -----------------------------------------------------------------------------
# 4) Fit Fractional Polynomial Model (Degree 3 with Fractional Powers)
# -----------------------------------------------------------------------------
model_frac = LinearRegression()
model_frac.fit(X_frac, y)

# Predict for a test matrix size (e.g., 300)
test_size = 300
X_test_frac = fractional_poly_transform([[test_size]], fractional_powers)
pred_frac = model_frac.predict(X_test_frac)[0]
print(f"\nPrediction for a matrix of size {test_size} using Fractional Polynomial (degree 3): {pred_frac:.2f}")

# -----------------------------------------------------------------------------
# 5) Plot the Data and Fractional Polynomial Fit
# -----------------------------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data (matrix size vs. iteration after max diff)')

# Create a smooth range of matrix sizes for plotting the fitted curve.
x_range = np.linspace(min(X), max(X), 100).reshape(-1, 1)
x_range_frac = fractional_poly_transform(x_range, fractional_powers)
y_frac = model_frac.predict(x_range_frac)

plt.plot(x_range, y_frac, color='red', label='Fractional Polynomial Fit (deg 3)')
plt.xlabel('Matrix Size')
plt.ylabel('Iteration after Maximum Error Change')
plt.title('Fractional Polynomial Model (Degree 3 with Fractional Powers)')
plt.legend()
plt.show()
```

```python
import torch
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  # For plotting and DataFrame creation

# Set device: Use "mps" if available (Apple M1), else fallback to "cpu"
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")

def newton_inverse(A, num_iterations=10, graph_error=False):
    """
    Compute the inverse of matrix A using Newton's method (Newton-Schulz iteration)
    in float32 precision and optionally graph the error at each iteration.
    
    Parameters:
        A (torch.Tensor): A square invertible matrix.
        num_iterations (int): Number of iterations to perform.
        graph_error (bool): If True, plot the error at each iteration.
        
    Returns:
        torch.Tensor: Approximation to the inverse of A.
        list: List of error norms at each iteration (length = num_iterations+1).
    """
    # Ensure A is on the correct device and in float32
    A = A.to(device).to(torch.float32)
    
    n, m = A.shape
    if n != m:
        raise ValueError("Input matrix A must be square.")
    
    # Create identity matrix in float32
    I = torch.eye(n, device=device, dtype=A.dtype)
    
    # Compute 1-norm (max column sum) and infinity-norm (max row sum)
    normA1 = torch.max(torch.sum(torch.abs(A), dim=0))
    normAinf = torch.max(torch.sum(torch.abs(A), dim=1))
    
    # Initial guess using the transposed matrix (in float32)
    X = A.t() / (normA1 * normAinf)
    
    errors = []
    initial_error = torch.norm(I - A @ X).item()
    errors.append(initial_error)
    
    for k in range(num_iterations):
        AX = A @ X
        correction = 2 * I - AX
        X = X @ correction
        error = torch.norm(I - A @ X).item()
        errors.append(error)
        # Try to clear unused memory (if on MPS)
        if device.type == "mps":
            torch.mps.empty_cache()
    
    if graph_error:
        plt.figure(figsize=(8, 5))
        plt.plot(range(num_iterations + 1), errors, marker='o', linestyle='-', color='b')
        plt.axhline(0, color='r', linestyle='--')
        plt.xlabel("Iteration")
        plt.ylabel("Error norm ||I - A*X||")
        plt.title("Error Norm vs. Iteration (float32)")
        plt.grid(True)
        plt.yscale("log")
        plt.show()
    
    return X, errors

# -------------------------------------------------------------------
# 1) Generate the data with 1 run per size & average errors
# -------------------------------------------------------------------
n_list = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, 4096]
max_iterations = 60
num_runs = 5  # Number of random matrices per size

data = []

for size in n_list:
    # Allocate tensor for errors in float32
    errors_all = torch.zeros((max_iterations + 1, num_runs), device=device, dtype=torch.float32)
    
    for run_idx in range(num_runs):
        # Generate a random matrix A with large entries in float32
        A = torch.ceil(1E3 * torch.rand(size, size, device=device, dtype=torch.float32))
        try:
            _, errors = newton_inverse(A, num_iterations=max_iterations, graph_error=False)
        except RuntimeError as e:
            print(f"Failed for size {size} at run {run_idx}: {e}")
            errors = [float('nan')] * (max_iterations + 1)
        errors_all[:, run_idx] = torch.tensor(errors, device=device, dtype=torch.float32)
    
    # Average the errors across runs for each iteration
    errors_mean = torch.mean(errors_all, dim=1)
    errors_mean = errors_mean.cpu().numpy()  # Convert to NumPy for plotting
    
    # Store the averages in our data structure
    for iteration, error_val in enumerate(errors_mean):
        data.append({'iteration': iteration, 'size': size, 'error': error_val})

df = pd.DataFrame(data)
print("Done")
print(df)
```

```python
import numpy as np

def format_matrix_scientific(matrix):
    """
    Returns a string representation of 'matrix' where each entry is formatted in scientific notation with 5 decimals.
    Handles both real and complex numbers.
    """
    rows = []
    for i in range(matrix.shape[0]):
        row_elems = []
        for j in range(matrix.shape[1]):
            x = matrix[i, j]
            # Use scientific notation to capture small numbers.
            if np.iscomplex(x):
                formatted = f"{x.real:.5e} + {x.imag:.5e}j"
            else:
                formatted = f"{x:.5e}"
            row_elems.append(formatted)
        rows.append("   ".join(row_elems))
    return "\n".join(rows)

def newton_inverse_to_file_no_arrows(A, iterations, output_filename, initial_guess=None):
    """
    Compute the inverse of matrix A using Newton's method and write each iteration's guess
    to a text file with each number formatted in scientific notation with 5 decimal places,
    using high precision arithmetic.
    If no initial guess is provided, X0 is computed as A.conj().T divided by the product of A's one norm and infinity norm.
    
    Parameters:
      A: A square numpy array with high-precision datatype (np.clongdouble).
      iterations: Number of Newton iterations.
      output_filename: The name of the file to which the output will be written.
      initial_guess: Optional initial guess for A^{-1}.
    
    Returns:
      The approximate inverse of A after the specified number of iterations.
    """
    # Ensure the matrix is square.
    if A.shape[0] != A.shape[1]:
        raise ValueError("Input matrix must be square.")
    
    n = A.shape[0]
    I = np.eye(n, dtype=A.dtype)
    
    # Compute the one norm (max absolute column sum) and infinity norm (max absolute row sum)
    if initial_guess is None:
        one_norm  = np.linalg.norm(A, 1) + 0.05e10
        inf_norm  = np.linalg.norm(A, np.inf) + 0.05e10 
        print("One norm:")
        print(one_norm)
        print("Inf norm:")
        print(inf_norm)
        # Use the conjugate transpose for complex matrices.
        initial_guess = A.T / ( one_norm * inf_norm)
    
    X = initial_guess.copy()
    
    with open(output_filename, 'w') as f:
        for i in range(iterations):
            f.write(f"Iteration {i}:\n")
            formatted_matrix = format_matrix_scientific(X)
            print(formatted_matrix)
            f.write(formatted_matrix + "\n\n")
            # Newton's iteration: X_{k+1} = X_k (2I - A @ X_k)
            X = X @ (2 * I - A @ X)
        f.write("Final approximation:\n")
        formatted_matrix = format_matrix_scientific(X)
        f.write(formatted_matrix + "\n")
    
    return X

# Define the test matrix (4x4 diagonal matrix with a complex entry) using high precision.
a = np.clongdouble(-17635117961.8806) + np.clongdouble(551276900.108857)*1j
A = np.array([
    [a, 0, 0, 0],
    [0, a, 0, 0],
    [0, 0, a, 0],
    [0, 0, 0, a]
], dtype=np.clongdouble)

# Run the function for a specified number of iterations (e.g., 30 iterations)
inverse_approx = newton_inverse_to_file_no_arrows(A, 30, "newton_iterations_no_arrows.txt")
```
