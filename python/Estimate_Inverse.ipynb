```python
import numpy as np
import matplotlib.pyplot as plt

def newton_inverse(A, num_iterations=10, graph_error=False):
    """
    Compute the inverse of matrix A using Newton's method (Newton-Schulz iteration)
    and optionally graph the error at each iteration, using the highest precision
    (np.longdouble) available.

    Parameters:
        A (numpy.ndarray): A square invertible matrix.
        num_iterations (int): Number of iterations to perform.
        graph_error (bool): If True, plot the error vs. iteration.

    Returns:
        numpy.ndarray: An approximation to the inverse of A.
    """
    # Convert A to an array with highest precision (np.longdouble)
    A = np.array(A, dtype=np.longdouble)

    # Check if A is square
    n, m = A.shape
    if n != m:
        raise ValueError("Input matrix A must be square.")
    
    # Create the identity matrix with highest precision
    I = np.eye(n, dtype=np.longdouble)
    
    # Initial guess for the inverse: X0 = A^T / (||A||_1 * ||A||_inf)
    normA1 = np.linalg.norm(A, 1)
    normAinf = np.linalg.norm(A, np.inf)
    X = A.T / (normA1 * normAinf)
    
    # List to store error at each iteration
    errors = []
    
    # Compute and store the initial error norm
    initial_error = np.linalg.norm(I - A @ X)
    errors.append(initial_error)
    
    print("Initial guess X0:")
    print(X)
    print("Initial error norm ||I - A*X||:", initial_error)
    print("-" * 50)
    
    # Newton-Schulz iteration: X_{k+1} = X_k (2I - A X_k)
    for k in range(num_iterations):
        # Compute A * X_k
        AX = A @ X
        
        # Compute the correction term: 2I - A * X_k
        correction = 2 * I - AX
        
        print(f"Iteration {k+1}:")
        print("A * X:")
        print(AX)
        print("Correction (2I - A * X):")
        print(correction)
        
        # Update the approximation
        X = X @ correction
        
        # Compute the current error norm ||I - A*X||
        error = np.linalg.norm(I - A @ X)
        errors.append(error)
        
        print("Updated X:")
        print(X)
        print(f"Error norm ||I - A*X||: {error}")
        print("-" * 50)
    
    # Graph the error if requested
    if graph_error:
        plt.figure(figsize=(8, 5))
        plt.plot(range(num_iterations + 1), errors, marker='o', linestyle='-', color='b', label='Error Norm')
        plt.axhline(0, color='r', linestyle='--', label='Zero Error')
        plt.xlabel("Iteration")
        plt.ylabel("Error norm ||I - A*X||")
        plt.title("Error Norm vs. Iteration")
        plt.grid(True)
        # Uncomment the following line if you prefer a logarithmic scale for the error:
        plt.yscale("log")
        plt.legend()
        plt.show()
    
    return X

# Example usage:
if __name__ == "__main__":
    A = np.array(result.tolist(),dtype=float)
    # Compute its inverse using 50 iterations of Newton's method and graph the error
    invA = newton_inverse(A, num_iterations=50, graph_error=True)
    
    print("Approximate inverse of A:")
    print(invA)
    print("Check A * invA (should be close to the identity matrix):")
    print(A @ invA)
```

```python
import torch
import pandas as pd
import numpy as np
from joblib import Parallel, delayed
import matplotlib.pyplot as plt
import gc  # For garbage collection

def generate_filled_powers(max_exp):
    """
    Generate a list of powers of two from 2^1 up to 2^(max_exp) and fill in the gaps
    between each consecutive power with the average of the two.

    Parameters:
        max_exp (int): The maximum exponent (e.g., 11 will generate up to 2^11).

    Returns:
        list: The list containing the powers of two with averages in between.
    """
    # Create the list of powers of two: [2, 4, 8, ..., 2^(max_exp)]
    powers = [2 ** i for i in range(1, max_exp + 1)]
    
    # Create a new list that will include the averages between consecutive values.
    filled_list = []
    for i in range(len(powers) - 1):
        filled_list.append(powers[i])
        # Calculate the average of the current and next power.
        avg = (powers[i] + powers[i+1]) // 2
        filled_list.append(avg)
    filled_list.append(powers[-1])
    
    # Print the resulting list
    print("Filled list:", filled_list)
    return filled_list



# Set device: use CUDA if available (for Nvidia GPUs on Windows 10), else fallback to CPU.
device = torch.device("mps")
print("Using device:", device)


# -------------------------------------------------------------------
# Parameters and Data Generation
# -------------------------------------------------------------------
n_list = [2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 1536, 2048, 4096]
#n_list =  generate_filled_powers(11)
#n_list = [16384]
max_iterations = 50
num_runs = 100  # Number of random matrices per size


def newton_inverse_torch(A, num_iterations=10):
    """
    Compute the inverse of matrix A using the Newton-Schulz iteration with PyTorch.
    The error at each iteration is computed.

    Parameters:
        A (torch.Tensor): A square invertible matrix (assumed to be float32).
        num_iterations (int): Number of iterations to perform.

    Returns:
        X (torch.Tensor): Approximation to the inverse of A.
        errors (list): List of error norms at each iteration.
    """

    print("Doing a matrix with dimensions: ", A.shape)    
    with torch.no_grad():
        n, m = A.shape
        if n != m:
            raise ValueError("Input matrix A must be square.")
        
        # Create identity matrix on the same device.
        I = torch.eye(n, device=device, dtype=A.dtype)
        
        # Compute norms for the initial guess.
        normA1 = torch.linalg.norm(A, 1)
        normAinf = torch.linalg.norm(A, float('inf'))
        
        # Initial guess: X0 = A^T / (||A||_1 * ||A||_inf)
        X = A.t() / (normA1 * normAinf)
        
        errors = []
        err = torch.linalg.norm(I - A @ X)
        errors.append(err.item())
        
        for k in range(num_iterations):
            AX = A @ X
            correction = 2 * I - AX
            X = X @ correction
            err = torch.linalg.norm(I - A @ X)
            errors.append(err.item())
        
        return X, errors

def compute_errors_for_size(size, num_iterations, num_runs):
    """
    Compute the average error curve (over multiple runs) for a given matrix size.
    Minimizes memory footprint by clearing GPU memory after each run.
    """
    errors_runs = []
    for _ in range(num_runs):
        with torch.no_grad():
            # Create a random matrix A on the device.
            A = torch.rand(size, size, device=device, dtype=torch.float16)
            _, errors = newton_inverse_torch(A, num_iterations=num_iterations)
            errors_runs.append(errors)
            # Delete A and clear cache to reduce memory usage.
            del A
            torch.cuda.empty_cache()
            gc.collect()
    # Compute the elementwise mean over runs.
    return np.mean(errors_runs, axis=0)


# Use joblib to sequentially process each matrix size.
results = Parallel(n_jobs=-1)(
    delayed(compute_errors_for_size)(size, max_iterations, num_runs) for size in n_list
)

# Build a list of dictionaries (one per iteration per matrix size)
data = [
    {'iteration': i, 'size': size, 'error': errors_mean[i]}
    for size, errors_mean in zip(n_list, results)
    for i in range(len(errors_mean))
]

df = pd.DataFrame(data)
print(df.head())

# -------------------------------------------------------------------
# Pickle the resulting DataFrame
# -------------------------------------------------------------------
pickle_filename = "df_results.pkl"
df.to_pickle(pickle_filename)
print(f"DataFrame successfully saved to {pickle_filename}")
```

```python
import numpy as np
import pandas as pd

# -------------------------------
# 1) Create the log-transformed DataFrame (only for error)
# -------------------------------
df_log = df.copy()

# Apply the natural logarithm only to the 'error' column.
df_log['log_error'] = np.log(df_log['error'])

# We keep the 'iteration' column as is.
# Sort by matrix size and then by iteration.
df_log = df_log.sort_values(['size', 'iteration']).reset_index(drop=True)

# -------------------------------
# 2) Apply the sliding window method on log(error) vs. iteration
# -------------------------------
epsilon = np.finfo(float).eps  # Small constant to avoid division by zero
window_size = 3

results_log = []  # To store the resulting point for each matrix size

# Process each group (each matrix size) separately.
for size, group in df_log.groupby('size'):
    group = group.reset_index(drop=True)
    
    # Ensure the group has enough points for a 5-point window.
    if len(group) < window_size:
        continue
    
    avg_log_derivatives = []  # List to hold the average derivative for each window.
    
    # Loop over each possible 5-point window in the current group.
    for i in range(len(group) - window_size + 1):
        window = group.iloc[i:i+window_size]
        # Use the normal iteration values for x and the log-transformed error for y.
        x = window['iteration'].values
        y = window['log_error'].values
        
        # Compute differences in x, avoiding division by zero.
        x_diff = np.diff(x)
        x_diff = np.where(np.abs(x_diff) < epsilon, epsilon, x_diff)
        slopes = np.diff(y) / x_diff
        
        # Compute the average derivative (slope) in this window.
        avg_slope = np.mean(slopes)
        avg_log_derivatives.append(avg_slope)
    
    # If no valid windows were found, skip this group.
    if len(avg_log_derivatives) == 0:
        continue
    
    # Find the index of the window with the most negative average derivative.
    # (i.e., the steepest drop in log(error) relative to iteration)
    min_window_index = np.argmin(avg_log_derivatives)
    
    # The rightmost point of a window of 5 points is at offset window_size - 1.
    rightmost_index = min_window_index + window_size - 1
    rightmost_point = group.iloc[rightmost_index]
    
    results_log.append({
        'size': size,
        'iteration': rightmost_point['iteration'],  # normal iteration value
        'log_error': rightmost_point['log_error'],      # log-transformed error
        'avg_log_slope': avg_log_derivatives[min_window_index]
    })

# Convert the results into a DataFrame.
result_df_log = pd.DataFrame(results_log)
print("Results on data with log-transformed error (iterations remain normal):")
print(result_df_log)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# --- Prepare the data ---
# Assume result_df_log has 'size' (matrix size) and 'iteration' (found iteration number)
X_fit = result_df_log[['size']].values  # matrix sizes (independent variable)
y_fit = result_df_log['iteration'].values  # iteration numbers (dependent variable)

# --- Log-transform the independent variable ---
X_log = np.log(X_fit)

# --- Fit a Linear Regression Model on the log-transformed data ---
model_log = LinearRegression()
model_log.fit(X_log, y_fit)

# Retrieve the fitted coefficients
a = model_log.coef_[0]       # slope
b = model_log.intercept_     # intercept

# Compute the predicted values on the training data (normal log fit)
yhat = model_log.predict(X_log)

# --- Adjust the model so it always sits above the data ---
# Compute the differences (if a data point is above the fitted curve, diff > 0)
diff = y_fit - yhat
max_diff = np.max(diff)

# Use a small margin to ensure the curve is strictly above the highest data point.
margin = 0.01
shift = max_diff + margin if max_diff > 0 else margin

# For the adjusted model, combine b and shift into a single intercept:
b_adjusted = b + shift

# --- Create a smooth range for plotting ---
x_range = np.linspace(result_df_log['size'].min(), result_df_log['size'].max(), 200).reshape(-1, 1)
x_range_log = np.log(x_range)

# Compute predictions for both fits on the smooth range.
y_range_normal = a * x_range_log + b           # Normal logarithmic fit
y_range_adjusted = a * x_range_log + b_adjusted   # Adjusted fit with combined intercept

# --- Plot the data and the two logarithmic fits side by side ---
fig, axs = plt.subplots(1, 2, figsize=(16, 6))

# Left subplot: Normal x-axis scale
axs[0].scatter(result_df_log['size'], result_df_log['iteration'], 
               color='blue', marker='o', label='Found Iteration')
axs[0].plot(x_range, y_range_normal, color='green', linewidth=2, 
            label='Normal Logarithmic Fit')
axs[0].plot(x_range, y_range_adjusted, color='red', linewidth=2, 
            label='Adjusted Logarithmic Fit')
axs[0].set_xlabel('Matrix Size')
axs[0].set_ylabel('Iteration Number')
axs[0].set_title('Normal X-Axis')
axs[0].legend()
axs[0].grid(True)

# Right subplot: Logarithmic x-axis scale
axs[1].scatter(result_df_log['size'], result_df_log['iteration'], 
               color='blue', marker='o', label='Found Iteration')
axs[1].plot(x_range, y_range_normal, color='green', linewidth=2, 
            label='Normal Logarithmic Fit')
axs[1].plot(x_range, y_range_adjusted, color='red', linewidth=2, 
            label='Adjusted Logarithmic Fit')
axs[1].set_xlabel('Matrix Size')
axs[1].set_ylabel('Iteration Number')
axs[1].set_title('Log X-Axis')
axs[1].legend()
axs[1].grid(True)
axs[1].set_xscale('log')

plt.tight_layout()

# --- Print the equations ---
print(f"Normal Logarithmic Fit: y = {a:.3f} * log(x) + {b:.3f}")
print(f"Adjusted Logarithmic Fit: y = {a:.3f} * log(x) + {b_adjusted:.3f}")

plt.show()
```

```python
import numpy as np
import pandas as pd
import math

def predict_iteration_ceiling_normal(x, a, b):
    """
    Predicts the iteration using the normal log fit and returns its ceiling.
    
    Equation: y = a * log(x) + b
    
    Parameters:
        x (float or array-like): Matrix size.
        a (float): Slope from the log fit.
        b (float): Intercept from the log fit.
    
    Returns:
        int or array-like: Ceiling of the predicted iteration.
    """
    pred = a * np.log(x) + b
    return int(np.ceil(pred)) if np.isscalar(x) else np.ceil(pred).astype(int)

def predict_iteration_ceiling_adjusted(x, a, b_adjusted):
    """
    Predicts the iteration using the adjusted log fit and returns its ceiling.
    
    Equation: y = a * log(x) + b_adjusted, where b_adjusted = b + shift.
    
    Parameters:
        x (float or array-like): Matrix size.
        a (float): Slope from the log fit.
        b_adjusted (float): Adjusted intercept (b + shift).
    
    Returns:
        int or array-like: Ceiling of the predicted iteration.
    """
    pred = a * np.log(x) + b_adjusted
    return int(np.ceil(pred)) if np.isscalar(x) else np.ceil(pred).astype(int)

# --- Create lookup table for powers of two only ---
# For example, using powers from 2^1 up to 2^63.
lookup_sizes = np.array([2 ** i for i in range(1, 64)])

# Create a DataFrame with predictions from both models.
lookup_table = pd.DataFrame({'matrix_size': lookup_sizes})
lookup_table['predicted_iteration_normal'] = lookup_table['matrix_size'].apply(lambda x: predict_iteration_ceiling_normal(x, a, b))
lookup_table['predicted_iteration_adjusted'] = lookup_table['matrix_size'].apply(lambda x: predict_iteration_ceiling_adjusted(x, a, b_adjusted))

print("Lookup Table:")
display(lookup_table)

# Define formatters for each column to prevent scientific notation.
formatters = {
    "matrix_size": lambda x: f"{x:.0f}",
    "predicted_iteration_normal": lambda x: f"{x:.0f}",
    "predicted_iteration_adjusted": lambda x: f"{x:.0f}"
}

# Convert the DataFrame to a formatted string.
formatted_table = lookup_table.to_string(index=False, formatters=formatters)

# Write the formatted string to a text file.
with open("lookup_table.txt", "w") as f:
    f.write(formatted_table)
```
